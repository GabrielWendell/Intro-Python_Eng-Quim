## 1. Criando nossos primeiros gráficos...

- Bibliotecas e *Magic lines*:

````python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
#%matplotlib notebook
````

- Gráfico mais simples possível:

````python
x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
````

````python
x = np.linspace(0,1,100)
y = np.random.randn(100)

plt.plot(x, y)

plt.xlabel('Tempo')
plt.ylabel('Valor')
````

- Customizando o estilo do gráfico:

````python
plt.plot(x, y, '.-', color = 'black', markersize = 20, alpha = 0.4, label = 'Gráfico 1')
plt.plot(x, y**(1/2), 'o', color = 'red', markersize = 15, alpha = 0.4, label = 'Gráfico 2')

plt.xlabel('Tempo')
plt.ylabel('Valor')
plt.title('Gráfico Randômico customizado')
plt.legend(loc = 0)


plt.grid()
plt.minorticks_on()
````

- Múltiplas figuras:

````python
plt.subplot(221)
plt.plot(x, y, 'k.')

plt.subplot(222)
plt.plot(x, y, 'y.')

plt.subplot(223)
plt.plot(x, y, 'r.')

plt.subplot(224)
plt.plot(x, y, 'g.')
````

- Figuras como objetos:

````python
fig = plt.figure(figsize = (8,5)) 
# ax1 = fig.add_axes([0.1,0.1,0.4,0.9])

ax1 = plt.subplot(111)
ax1.plot(x, y)
ax1.set_xlabel('Tempo')
ax1.set_ylabel('Valor')
````

## 2. Customizando gráficos

- Estilos estéticos:

````python
def sm_style():
    # plt.style.use('dark_background')
    plt.grid(True, linestyle = ':', color = '0.50')
    plt.rcParams['font.family'] = 'monospace'
    plt.rcParams['font.size'] = 15
    plt.minorticks_on()
    plt.tick_params(axis = 'both', which = 'minor', direction = "in",
                        top = True, right = True, length = 5, width = 1, labelsize = 15)
    plt.tick_params(axis = 'both', which = 'major', direction = "in",
                        top = True, right = True, length = 8, width = 1, labelsize = 15)
````

````python
def style():
    from matplotlib import rcParams
    import matplotlib.font_manager as font_manager
    
    rcParams['font.family'] = 'monospace'
    rcParams['mathtext.fontset'] = 'dejavuserif'
    rcParams['axes.linewidth'] = 1.2
    
style()
````

- Plot completo:

````python
fig = plt.figure(figsize = (10, 10))

ax1 = plt.subplot(2,1,1)
sm_style()
ax2 = plt.subplot(2,2,3)
sm_style()
ax3 = plt.subplot(2,2,4)
sm_style()

ax1.plot(x, y, '.-', color = 'k')
ax2.plot(x, y, '.-', color = 'r')
ax3.plot(x, y, '.-', color = 'g')

ax1.set_xlabel('tempo', {'color': 'k', 'fontsize': 20})
ax2.set_xlabel('tempo', {'color': 'k', 'fontsize': 20})
ax3.set_xlabel('tempo', {'color': 'k', 'fontsize': 20})

ax1.set_ylabel('Valor', {'color': 'k', 'fontsize': 20})
ax2.set_ylabel('Valor', {'color': 'k', 'fontsize': 20})


ax1.set_ylim(-5, 5)

ax1.minorticks_on()
ax2.minorticks_on()
ax3.minorticks_on()

ax1.set_title('Teste', size = 20)


plt.savefig('Teste.png', dpi = 150)
````

- Plots compostos:

````python
fig = plt.figure(figsize = (10, 5)) 
ax1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax1.plot(x, y)
sm_style()

ax1.set_ylim(-5, 8)

ax2 = fig.add_axes([0.5, 0.65, 0.3, 0.3])
ax2.plot(x, y, 'k-')
ax2.set_xlim(0.4, 0.6)
ax2.set_title('Zoom', size = 10)
sm_style()
````

- Plots múltiplos:

````python
fig, axes = plt.subplots(figsize = (8,5), nrows = 2, ncols = 2)

axes[0,0].plot(x, y, 'k-')
axes[0,1].plot(x, y, 'r-')
axes[1,0].plot(x, y, 'b-')
axes[1,1].plot(x, y, 'g-')
````

- Estilos diferentes:

````python
fig = plt.figure(figsize = (8,5))
ax1 = plt.subplot(111)

plt.plot(x, y, marker = '.', color = 'blue', lw = 1, ls = '--', markersize = 15,
        markerfacecolor = 'red', markeredgecolor = 'black', markeredgewidth = 3)

plt.xlabel('Tempo', {'color': 'red', 'fontsize': 20, 'alpha': 0.5,
                    'fontstyle':'italic'})


plt.grid(True, color = 'green', alpha = 0.5, dashes = (5,1))

ax1.set_facecolor('0.9')

sm_style()
````

## 3. Diferentes tipos de gráficos:

- Histogramas:

````python
def gaussian(x, amp, cen, larg):
    return (amp / (np.sqrt(2*np.pi) * larg)) * np.exp(-(x-cen)**2 / (2*larg**2))

x = np.linspace(1,100,100)
y = gaussian(x,2,50,20)


plt.bar(x,y)
sm_style()
````

````python
fig = plt.figure(figsize = (10,5))
plt.bar(x,y,width = 0.8)
sm_style()
````

````python
fig = plt.figure(figsize = (10,5))
plt.hist(y, bins = 20, density = True, histtype = 'step')
sm_style()
````

````python
x = np.random.rand(100)

plt.hist(x)
sm_style()
````

- Histograma de um gaussiano real:

````python
prob1 = np.random.randint(1,7,1000)

plt.hist(prob1, bins = 6, histtype = 'step', color = 'r')
sm_style()
````

````python
prob1 = np.random.randint(1,33,500)
prob2 = np.random.randint(1,33,500)

prob3 = prob1 + prob2

plt.hist(prob3, bins = 31, histtype = 'step', color = 'r')
sm_style()
````

- Gráfico de barra:

````python
x = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jun', 'Ago', 'Set']
y = [5, 10, 8, 12, 20, 25, 28, 25, 22]
var = [1, 2, 1, 3, 2, 5, 1, 3, 2]

plt.bar(x, y, yerr = var, color = 'blue', width = 0.5)
sm_style()
````

````python
x = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jun', 'Ago', 'Set']
y = np.array([5, 10, 8, 12, 20, 25, 28, 25, 22])

var = [1, 2, 1, 3, 2, 5, 1, 3, 2]
Ox = np.arange(9)

plt.bar(Ox, y, yerr = var, color = 'blue', width = 0.5)
plt.bar(Ox + 0.5, y/2, yerr = var, color = 'red', width = 0.5)
plt.xticks(Ox, x)

sm_style()
````

- Plots de pontos (*scatter plots*):

````python
Prot = np.linspace(1, 50, 30)
Prot = Prot + Prot*(np.random.rand(30)/10)
Ali = np.linspace(5,0.8,30)
Ali = Ali + Ali*(np.random.rand(30)/10)
Age = np.linspace(0.6,9,30)
Age = Age + Age*(np.random.rand(30)/1)

fig = plt.figure(figsize = (5,5))
plt.scatter(Prot, Ali)
sm_style()
````

````python
fig = plt.figure(figsize = (5,5))
plt.scatter(Prot, Ali, s = Age*20, c = Prot)
plt.subplots_adjust(right = 0.88)

plt.colorbar()
sm_style()
````
